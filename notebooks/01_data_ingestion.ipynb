# Load IBM AMLSim data

```python
urls = {
       'accounts': 'https://raw.githubusercontent.com/Bidhisha24/next-generation-aml-investigation-framework/refs/heads/main/data/IBM%20AMLSim/accounts.csv',
       'alerts': 'https://raw.githubusercontent.com/Bidhisha24/next-generation-aml-investigation-framework/refs/heads/main/data/IBM%20AMLSim/alerts.csv',
       'transactions': 'https://raw.githubusercontent.com/Bidhisha24/next-generation-aml-investigation-framework/refs/heads/main/data/IBM%20AMLSim/transactions.csv'
    }
for data in urls.keys():
    url = urls[data]
    df = pd.read_csv(url)
    if data == 'accounts':
        accounts = spark.createDataFrame(df)
    elif data == 'alerts':
        alerts = spark.createDataFrame(df)
    elif data == 'transactions':
        transactions = spark.createDataFrame(df)
print("Accounts data:")
display(accounts)
print("Alerts data:")
display(alerts)
print("Transactions data:")
display(transactions)
```

# Load SAML-D data

```python
saml_d = spark.table("aml_framework.transactions")
saml_d.count(), len(saml_d.columns)
```

```python
accounts.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("aml_framework.accounts")
alerts.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("aml_framework.alerts")
transactions.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("aml_framework.transactions")
saml_d.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("aml_framework.saml_d")
```
